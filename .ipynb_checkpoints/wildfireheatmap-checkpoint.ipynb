```python
import pandas as pd
import gzip
```

```python
df = pd.read_csv('wildfiredataset.gz', compression='gzip')
```

```python
df
```

```python
df.PROVINCE.value_counts()
```

```python
df[['YEAR','MONTH','etc']] = df.ACQ_DATE.str.split("-",expand=True,)
df
```

```python
df.PROVINCE.value_counts()
```

```python
dfnew = df.groupby(["YEAR", "PROVINCE"]).size().reset_index(name="FREQ")
```

```python
from plotnine import *
from plotnine.data import mpg
%matplotlib inline


white = '#fff'

ggplot(data = dfnew, mapping = aes(x = 'PROVINCE',
                                                   y = 'YEAR',
                                                   fill = 'FREQ', color = 'white')) + geom_tile() + theme(axis_text_x=element_text(angle=90)) + theme(figure_size=(12, 6))  

```
