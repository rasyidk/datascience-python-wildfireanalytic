```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gzip
```

```python
df = pd.read_csv('wildfiredataset.gz', compression='gzip')
```

```python
df
```

```python
df[['YEAR','MONTH','etc']] = df.ACQ_DATE.str.split("-",expand=True,)
df
```

```python
dfnew = df.groupby(["palm_oilt"]).size().reset_index(name="FREQ")
```

```python
dfnew
```

```python
dfnew.head(10)
```

```python
y_pos = np.arange(len(dfnew.palm_oilt))
 
# Create bars
plt.bar(y_pos, dfnew.FREQ)
 
# Create names on the x-axis
plt.xticks(y_pos, dfnew.palm_oilt)
 
# Show graphic
plt.show()
```
